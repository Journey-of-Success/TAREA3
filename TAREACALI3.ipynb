```python
# link to repo

dataRepo = 'https://github.com/Journey-of-Success/TAREA3/raw/main/appendData/'
```

```python
# creating file names into a list:
numbers=range(1,5)
fileNames=['level'+str(number)+'.xlsx' for number in numbers]

# list of file names
fileNames
```

```python
# creating the url to each file:
alltheLinks=[dataRepo+fn for fn in fileNames]
alltheLinks
```

```python
#to save each data in a list allDFs we need openpyxl and xlrd for excel
!pip install openpyxl
```

```python
# creating list of DFs
import pandas as pd

allDFs=[pd.read_excel(link,engine='openpyxl') for link in alltheLinks]
```

```python
allDFs[0].columns=allDFs[0].iloc[3, :].tolist()
allDFs[0].head()
```

```python
column_names= allDFs[0].columns.tolist()
column_names
```

```python
# a√±adir un contador para contar 'HDI rank'
count = 0

# Iterando la lista para contar
for i, col in enumerate(column_names):
    if col == 'HDI rank':
        count += 1
        # si aparece por segunda vez, renombrar por 'HDI rank.1'
        if count == 2:
            column_names[i] = 'HDI rank.1'
            break

column_names
```

```python
allDFs[0].columns=column_names
allDFs[0].head()
```

```python
#funcion para eliminar filas no necesarias
def clean_rows(df, column_name):
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
        cleaned_df = df.dropna(subset=[column_name])
        return cleaned_df
```

```python
# Cleaning each Data Frame using the function
allDFs = [clean_rows(df, 'HDI rank') for df in allDFs]
```

```python
# saving column names
allColumnNames=[]
for df in allDFs:
    allColumnNames.append(set(df.columns))
```

```python
# checking how many columns per df
[len(cols) for cols in allColumnNames]
```

```python
# details of common columns
commonColumns=set.intersection(*allColumnNames)
len(commonColumns),commonColumns
```

```python
# all minus the common
set.union(*allColumnNames)-commonColumns
```

```python
# DFs with the common columns
allDFs_sameNames=[df.loc[:,list(commonColumns)] for df in allDFs]
allDFs_sameNames=[df.reset_index(drop=True) for df in allDFs_sameNames]
```

```python
# appending
allDFsConcat=pd.concat(allDFs_sameNames)
allDFsConcat
```

```python
#para ordenar
#usamos inplace para guardar los datos (modificar)
allDFsConcat.set_index(['HDI rank', 'Country'],inplace=True)
allDFsConcat.head()
```

```python
# ordering column names alphabetically
allDFsConcat.sort_index(axis=1,inplace=True) # by row index will be axis=0
allDFsConcat
```

```python
# indexes will be columns
allDFsConcat.reset_index(inplace=True)
allDFsConcat
```

```python
import pandas as pd

fallecimientos = 'https://raw.github.com/Journey-of-Success/TAREA3/main/fallecidosCOVID.csv'
# Leer la tabla
df = pd.read_csv(fallecimientos)
df.head()
```

```python
save_columns=df.columns.str.split(';')
saving=save_columns[0]
```

```python
df.rename(columns={df.columns[0]:'a'},inplace=True)
df=df.a.str.split(';',expand=True)
df.info()
```

```python
new={0:saving[0],1:saving[1],2:saving[2],3:saving[3],4:saving[4],5:saving[5],6:saving[6],7:saving[7],8:saving[8],9:saving[9]}
df.rename(columns=new,inplace=True)
df.info()
```

```python
df.head()
```

```python
# Convertir la columna de fecha a tipo datetime
df['FECHA_FALLECIMIENTO'] = pd.to_datetime(df['FECHA_FALLECIMIENTO'])
df.head()
```

```python
# Contar el n√∫mero de registros por a√±o y mes
df['conteo'] = 1

# Agrupar por a√±o y mes y sumar los conteos
result = df.groupby(['a√±o','mes']).agg({'conteo':'sum'}).reset_index()

# Mostrar el resultado
print(result)
```
